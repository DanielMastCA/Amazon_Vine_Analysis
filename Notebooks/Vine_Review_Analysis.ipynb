```python
import pandas as pd
path = '../Resources/Vine_Data.csv'
df = pd.read_csv(path)
df
```

```python
filtered_df = df.loc[df["total_votes"] >= 20]
filtered_df
```

```python
filtered_df = filtered_df.loc[(filtered_df["helpful_votes"] / filtered_df["total_votes"] * 100) >= 0.5]
filtered_df
```

```python
vine_df = filtered_df.loc[filtered_df['vine'] == 'Y']
vine_df
```

```python
unpaid_df = filtered_df.loc[filtered_df['vine'] == 'N']
unpaid_df
```

```python
total_reviews = filtered_df.count()
print("There are " + str(total_reviews["verified_purchase"]) + " total reviews")
```

```python
five_stars = filtered_df.loc[filtered_df["star_rating"] == 5].count()

print("There are " + str(five_stars["star_rating"]) + " 5 star reviews")
```

```python
unpaid_five_stars = unpaid_df.loc[filtered_df["star_rating"] == 5].count()

print(str(unpaid_five_stars["star_rating"] / five_stars["star_rating"]) + "% of the 5 star ratings are unpaid")
```

```python
paid_five_stars = vine_df.loc[filtered_df["star_rating"] == 5].count()

print(str(paid_five_stars["star_rating"] / five_stars["star_rating"]) + "% of the 5 star ratings are paid")
```
